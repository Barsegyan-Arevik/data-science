```python
import re
import string

# Функция фильтрации
def text_filtering(text):
    # Избавляемся от спец. символов
    text = text.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ').replace('\'', '').replace("'", '')
    
    # Удаляем знаки пунктуации
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Удаляем все символы не относящиеся к кириллице
    text = re.sub("[^a-zA-Z ]", "", text)
    
    # Заменяем множественные пробелы
    text = re.sub("\s{2,}", " ", text)
    
    # Удаляем пробелы по бокам
    text = text.strip()
    
    # Возвращаем отфильтрованный текст
    return text
```

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords

for i in nltk.word_tokenize(text):
    newText = lemma.lemmatize(i, get_wordnet_pos(i))
    print(newText)
    tokens.append(newText)
    

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)


def text_lemmatizer(text):
    lemma = nltk.WordNetLemmatizer()
    # Приводим весь текст к нижнему регистру
    text = text.lower()
    # Разбиваем текст на слова
    qwords = text.split()
    
    # Вводим лист стоп-слов
    stop_words = stopwords.words('english')

    # Получаем токены
    tokens = []
    for word in qwords:
        
        if word in correct_words:
            word_in_normal_form = lemma.lemmatize(word, get_wordnet_pos(word)) 
        else:
            word = typo_correction(word)
            word_in_normal_form = lemma.lemmatize(word, get_wordnet_pos(word))
        
        if word_in_normal_form not in stop_words and word_in_normal_form in correct_words:
            tokens.append(word_in_normal_form)

    # Объединяем токены в текст
    text = ' '.join(tokens)
    
    # Возвращаем лемматизированный текст
    return text

```

```python
import torch
from transformers import AutoModelForSequenceClassification
from transformers import BertTokenizerFast

# Получаем токены из предобученной модели
tokenizer = BertTokenizerFast.from_pretrained('blanchefort/rubert-base-cased-sentiment')
# Подгружаем предобученную модель
model = BertTokenizerFast.from_pretrained('blanchefort/rubert-base-cased-sentiment', return_dict=True)

# Вызов декоратора для отключения вычислений градиента (мы не обучаем и не оптимизируем)
@torch.no_grad()
def predict(text):
    # Создаём словарь эмоциональности
    emotions_dict = {"NEUTRAL": 0, "POSITIVE": 0, "NEGATIVE": 0}
    # Инициализируем переменную предсказания
    result = ""
    
    '''
    text - это наш текст
    max_length - максимальная длина (колличество токенов)
    padding - заполняет короткие последовательности
    truncation - усекает длинные последовательности
    return_tensors - возвращает объект структуры тензор. объект в векторном пространстве
    '''

    # Вводим входные параметры модели
    inputs = tokenizer(text, max_length=512, padding=True, truncation=True, return_tensors='pt')
    # Выводим результат модели
    '''
    **inputs - Упаковываем в словарь все входные параметры
    '''
    outputs = model(**inputs)
    # Применяем функцию softmax для наших результатов (масштабируем результат)
    '''
    logits - результат предсказания
    '''
    predicted = torch.nn.functional.softmax(outputs.logits, dim=1)
    # Заполняем словарь значениями
    emotions_dict["NEUTRAL"] = round(float(predicted[0][0].numpy()) * 100, 2)
    emotions_dict["POSITIVE"] = round(float(predicted[0][1].numpy()) * 100, 2)
    emotions_dict["NEGATIVE"] = round(float(predicted[0][2].numpy()) * 100, 2)
    # Возвращаем индекс максимального значения
    predicted = torch.argmax(predicted, dim=1).numpy()
    # Записываем результат предсказания в переменную
    result = max(emotions_dict, key=emotions_dict.get)
    return emotions_dict, result
  

```

```python
import pandas as pd

# Создаём листы, чтобы в будущем, использовать их как значения колонок
review = []
processed_text = []
emotions = []
results = []

# Создаём переменные, для записи результатов нейронной сети
emotions_dict = {}
result = ""

# Чтение данных из табличной структуры
dataframe = pd.read_csv(path + "test.csv")

# Циклично проходимся по всей колонке
for val in dataframe["tweet"]:
    # Предобработка текста
    text = text_preprocessing(str(val))
    if not text:
        continue
    # Получение предсказания эмоциональности
    emotions_dict, result = predict(text)
    # Запись результатов в списки
    review.append(val)
    processed_text.append(text)
    emotions.append(emotions_dict)
    results.append(result)
    
# Инициализация табличной структуры
emotion_df = pd.DataFrame()
# Запись результатов в табличную структуру
emotion_df["Review"] = review
emotion_df["Processed text"] = processed_text
emotion_df["Emotions"] = emotions
emotion_df["Results"] = results
# Вывод табличной структуры
emotion_df
```
